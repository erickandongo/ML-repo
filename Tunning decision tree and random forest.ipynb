# One Hot Encoding

```python
df.info()
df_dummy = pd.get_dummies(df, prefix_sep='_', drop_first=True)
```

```python
final_tr = pd.DataFrame(data=df_dummy)

print(final_tr.shape)
```

# Train Test Split

```python
# set x and y variables
y = final_tr['isFraud']
x= final_tr.loc[:, final_tr.columns != 'isFraud']


from sklearn.model_selection import train_test_split

# Split randomly into 80% train data and 20% test data
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 0.2, random_state = 123)
```

```python
from sklearn.utils import resample

# Concatenate 'final_tr' and 'yTrain' along the columns axis
data = pd.concat([xTrain, yTrain], axis=1)

# Separate the majority and minority classes
majority_class = data[data['isFraud'] == 0]  # Assuming the majority class is labeled as 0
minority_class = data[data['isFraud'] == 1]  # Assuming the minority class is labeled as 1

# Upsample the minority class to match the majority class
minority_upsampled = resample(minority_class, 
                              replace=True,     # Sample with replacement
                              n_samples=len(majority_class),   # Match the number of samples in the majority class
                              random_state=42)  # Set a random seed for reproducibility

# Combine the upsampled minority class with the majority class
resampled_data = pd.concat([majority_class, minority_upsampled])

# Separate the features (X) and labels (y) from the resampled data
X_resampled = resampled_data.drop('isFraud', axis=1)
y_resampled = resampled_data['isFraud']

# Print the resampled data
print(X_resampled.head())
print(y_resampled.head())
```

# Build Decision Tree Model with Hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
#Using max_depth, criterion will suffice for DT Models, rest all will remain constant 
parameters = {'max_depth' : (3,5,7,9,10,15,20,25)
              , 'criterion' : ('gini', 'entropy')
              , 'max_features' : ('auto', 'sqrt', 'log2')
              , 'min_samples_split' : (2,4,6)
             }
```

```python
DT_grid  = RandomizedSearchCV(DecisionTreeClassifier(), param_distributions = parameters, cv = 5, verbose = True) 
```

```python
DT_grid.fit(X_resampled,y_resampled)
```

```python
DT_grid.best_estimator_
```

```python
#Re Build Model with Best Estimators
DT_Model = DecisionTreeClassifier(criterion='entropy',
                       max_depth=20, max_features='sqrt')
```

```python
Model = DT_grid.fit(X_resampled,y_resampled)
```

```python
predictions = DT_grid.predict(xTest)
predictions_train = DT_grid.predict(X_resampled)
```

```python
train_accuracy = DT_grid.score(X_resampled, y_resampled)
test_accuracy = DT_grid.score(xTest, yTest)
```

```python
print(f'Train Accuracy: {train_accuracy:.3f}')
print(f'Test Accuracy: {test_accuracy:.3f}')
```

# Build Random Forest Model with hyperparameters

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 80, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [2,4]
# Minimum number of samples required to split a node
min_samples_split = [2, 5]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2]
# Method of selecting samples for training each tree
bootstrap = [True, False]
```

```python
# Create the param grid
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(param_grid)
```

```python
rf_Model = RandomForestClassifier()
```

```python
rf_Grid = GridSearchCV(estimator = rf_Model, param_grid = param_grid, cv = 3, verbose=2, n_jobs = 4)
```

```python
rf_Grid.fit(X_resampled,y_resampled)
```

```python
rf_Grid.best_params_
```

```python
print (f'Train Accuracy - : {rf_Grid.score(X_resampled,y_resampled):.3f}')
print (f'Test Accuracy - : {rf_Grid.score(xTest,yTest):.3f}')
```
